```python
#importing the libraries
import pandas as pd
```

```python
#importing dataset
mobile_dataset = pd.read_csv('Mobile_Price.csv')

#Remove rows with missing values
mobile_dataset = mobile_dataset.dropna(axis=0)
```

```python
#Set dependent and independent variables
y = mobile_dataset.price_range
X = mobile_dataset.iloc[:,:-1]
```

```python
#Splitting the dataset into training set and testing set
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X,y, random_state=0)
```

```python
#import Decision Tree Regressor and Mean Absolute Error
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error

#define a function to get the mean absolute value for each number of max leaf nodes
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    mobile_model = DecisionTreeRegressor(max_leaf_nodes = max_leaf_nodes, random_state=0)
    mobile_model.fit(train_X, train_y)
    val_predicted = mobile_model.predict(val_X)
    mae = mean_absolute_error(val_y, val_predicted)
    return (mae)
```

```python
#find the best number of max_leaf_nodes using for loop
best_mae = float('inf')
best_tree_size = None
for max_leaf_nodes in [100, 101, 102, 103, 104, 105 ,106, 107, 108, 109, 110, 111, 112, 113, 114, 115]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    if my_mae < best_mae:
        best_mae = my_mae
        best_tree_size = max_leaf_nodes

print(f"Best tree size: {best_tree_size}")
print(f"Least MAE: {best_mae}")
```

```python
#get the predictions using best_tree_size
#define the model
best_model = DecisionTreeRegressor(max_leaf_nodes = best_tree_size ,random_state=0)

#fit the model
best_model.fit(train_X, train_y)

#predict mobile price from val_X
val_predicted = best_model.predict(val_X)
print(val_predicted[:5])
```

```python
#Random Forest
#Many decision trees voting together
from sklearn.ensemble import RandomForestRegressor

#define the model
rf_model = RandomForestRegressor(random_state = 0)
#fit the model
rf_model.fit(train_X, train_y)

#predict mobile price from val_X
val_rf_predicted = rf_model.predict(val_X)
print(val_rf_predicted[:5])

#print MAE
print(mean_absolute_error(val_y, val_rf_predicted))
```
